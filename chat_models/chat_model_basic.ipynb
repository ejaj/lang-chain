```python
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
```

```python
load_dotenv()
```

```python
# Create a ChatOpenAI model
model = ChatOpenAI(model="gpt-4o")
```

```python
# Invoke the model with a message
result = model.invoke("What is your age?")
print(result)
```
